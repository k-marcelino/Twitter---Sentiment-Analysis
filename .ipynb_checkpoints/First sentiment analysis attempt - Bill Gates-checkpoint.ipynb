```python
# Importing libraries and packages
import snscrape.modules.twitter as sntwitter
import pandas as pd
import time
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np

from textblob import TextBlob
from wordcloud import WordCloud
import re
plt.style.use('fivethirtyeight')
```

```python
start_time = time.time()

# Creating list to append tweet data 
tweets_list1 = []


# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('from:BillGates').get_items()): #declare a username 
    if i>100000: #number of tweets you want to scrape
        break
    #declare the attributes to be returned
    tweets_list1.append([tweet.date,  
                         tweet.content, 
                         tweet.source, 
                         tweet.user.username, 
                         tweet.media,
                         tweet.lang,
                         tweet.user.verified,]) 
    
# Creating a dataframe from the tweets list above 
bill = pd.DataFrame(tweets_list1, columns=['Datetime',
                                                 'Text',
                                                 'Source',
                                                 'Username',
                                                 'Media',
                                                 'language',
                                                 'verified'])

end_time = time.time()
print(f"Result calculated in {end_time - start_time} sec")
```

```python
# Cleaning text > 15:50 do vídeo

def cleantxt(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) #Remove @mentions
    text = re.sub(r'#', '', text) # Removing #
    text = re.sub(r'RT[\s]+', '', text) #Removing RT
    text = re.sub(r'https?:\/\/\S+', '', text) # Removing hiperlink
    
    return text
```

```python
# Using ".apply"

bill['clean_tweets'] = bill['Text'].apply(cleantxt)
bill['clean_tweets']
```

```python
# Using ".agg"

bill['clean_tweets2'] = bill['Text'].agg(cleantxt)
bill['clean_tweets2']
```

```python
# Function do get subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

# Function to get polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
# Creating new columns

bill['Subjectivity'] = bill['clean_tweets'].apply(getSubjectivity)
bill['Polarity'] = bill['clean_tweets'].apply(getPolarity)

bill[['clean_tweets','Subjectivity', 'Polarity']]
```

```python
# Plot the wordcloud

allWords = ' '.join([twts for twts in bill['clean_tweets']])
wordCloud = WordCloud(width=500, height=300, random_state=21, max_font_size=119).generate(allWords)

plt.imshow(wordCloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

```python
# Function do compute negative, neutral and positive analysis

def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
bill['Analysis'] = bill['Polarity'].apply(getAnalysis)
bill[['clean_tweets','Subjectivity', 'Polarity', 'Analysis']]
```

```python
# Print positive tweets

bill.sort_values(['Polarity','Subjectivity'], ascending=[True, True])
```

```python
# Plotting the polarity ans subjectivity

plt.figure(figsize=(8,6))
for i in range(0, bill.shape[0]):
    plt.scatter(bill['Polarity'][i], bill['Subjectivity'][i], color='Blue')
    
plt.title('Sentiment Analysis')
plt.xlabel('Polarity')
plt.ylabel('Subjectivity')
plt.show()
```

```python
# Get the percentage of positive, negative and neutral tweets

#ptweets = bill[bill.Analysis == 'Positive']
ptweets = bill[bill['Analysis'] == 'Positive'].shape[0]
ntweets = bill[bill['Analysis'] == 'Negative'].shape[0]
neutral = bill[bill['Analysis'] == 'Neutral'].shape[0]
print(f'Number of positive tweets: {ptweets}')
print(f'Number of negative tweets: {ntweets}')
print(f'Number of neutral tweets: {neutral}\n')

print(f'Percentagem of positive tweets: {ptweets / bill.shape[0]:.2f}')
print(f'Percentagem of negative tweets: {ntweets / bill.shape[0]:.2f}')
print(f'Percentagem of neutral tweets: {neutral / bill.shape[0]:.2f}')
```
